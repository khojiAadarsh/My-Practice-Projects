### Using Pivot to group the data 
Grouping on the basis of Type of Crime(Group_Name) and the year(Year).

```python
groupwise = pd.pivot_table(crimedata , values = ["Cases_Chargesheeted" , "Cases_Convicted", "Cases_Acquitted_or_Discharged"],
                          index = ["Group_Name", "Year"], aggfunc = sum)
groupwise
```

```python
statewise = pd.pivot_table(crimedata , values = ["Cases_Chargesheeted"],
                          index = ["Area_Name", "Year","Group_Name"], aggfunc = sum)
statewise
```

### Using loc/iloc for Slicing in Pandas 

```python
crime_MP_2001 = crimedata.loc[(crimedata.Area_Name == 'Madhya Pradesh') & (crimedata.Year == 2001)]
```

```python
display(crime_MP_2001)
```

### Plotting a bargraph using Matplotlib and Seaborn
Bargraph to visualise No of Crimes across all groups in Madhya Pradesh for the year 2001.

```python
fig = plt.figure(figsize=(20,9))
plt.title("No of crimes across groups in MP for 2001" , size =25)
ax = sns.barplot(data = crime_MP_2001.iloc[:100], y = "Cases_Chargesheeted", x = "Group_Name", linewidth = 2,edgecolor = 'red')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```

```python
rape_MP = crimedata.loc[(crimedata.Group_Name == 'Rape') & (crimedata.Area_Name == 'Madhya Pradesh')]
```

```python
display(rape_MP)
```

Bargraph to visualise No of Chaargesheets filed for Rape cases in Madhya Pradesh from 2001 to 2010.

```python
fig = plt.figure(figsize=(16,9))
plt.title("No of Rape chargesheets across 10 years in Madhya Pradesh" , size =25)
ax = sns.barplot(data = rape_MP.iloc[:100], y = "Cases_Chargesheeted", x = "Year", linewidth = 2,edgecolor = 'red')

plt.show()
```

```python
Dowrydeathcases = crimedata.loc[(crimedata.Group_Name == 'Dowry Deaths')]
```

```python
display(Dowrydeathcases)
```

### Generating a Heatmap using Matplotlib and Seaborn
Heatmap to visualise the Regional comparison of Reported Rape cases(Chargesheeted) between 2001 and 2010 for all States and Union Territories

```python
plt.figure(figsize=(16, 9))
sns.heatmap(Dowrydeathcases.pivot_table(values='Cases_Chargesheeted', index='Area_Name', columns='Year', aggfunc='sum'), cmap='YlGnBu')
plt.title('Regional Comparison of Reported Rape Cases Over Years')
plt.xlabel('Year')
plt.ylabel('Area Name')
plt.show()
```

### Generating a Scatterplot using Matplotlib and Seaborn
Scatterplot for the Correlation Analysis between No of cases Chargesheeted and No of Cases with conviction in Madhya Pradesh

```python
Dowrydeathcases_MP = Dowrydeathcases.loc[(crimedata.Area_Name == 'Madhya Pradesh')]
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Cases_Chargesheeted', y='Cases_Convicted', data=Dowrydeathcases_MP)
plt.title('Correlation Analysis between No of cases chargesheeted and no of convicted cases')
plt.xlabel('No of Cases chargesheeted')
plt.ylabel('No of Cases with conviction')
plt.show()
```
